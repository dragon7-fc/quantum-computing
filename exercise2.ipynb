## 2: Black-body radiation

$I_\lambda(\lambda, T) = \frac{2hc^2}{\lambda^5}\frac{1}{e^{\frac{hc}{\lambda kT}} - 1}$

* $\lambda: wave\ length$
* $T: Temperature$
* $h: Planck\ constant = 6.62607004x10^{-34}$
* $k: Boltzmann\ constant = 1.38064852x10^{-23}$
* $c: speed of the light = 3x10^8$

![black-body_radiation](black-body_radiation.png)

```python
# author: FC Su
# copyright: Copyright (C) 2020 FC Su
# license: Public Domain
# version: 1.0
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

h = 6.62607004e-34  # Planck constant
k = 1.38064852e-23  # Boltzmann constant
c = 3e8  # speed of the light

# intensity
def I(wavelength, T):
    a = 2*h * c**2/wavelength**5
    b = 1/(np.exp((h*c)/(wavelength*k*T)) -1)
    return a*b

wavelengths = np.arange(1e-9, 2e-6, 1e-9)
intensity3k = I(wavelengths, 3000)
intensity4k = I(wavelengths, 4000)
intensity5k = I(wavelengths, 5000)
intensity6k = I(wavelengths, 6000)
intensity7k = I(wavelengths, 7000)

plt.plot(intensity3k)
plt.plot(intensity4k)
plt.plot(intensity5k)
plt.plot(intensity6k)
plt.plot(intensity7k)
plt.xlabel('wavelength(nm)')
plt.ylabel('spectral irradiance')
plt.title('Blackbody Radiation')
plt.legend(['3000', '4000', '5000', '6000', '7000'])
plt.grid()
plt.show()
```

## 3: Generate a song

![happy_birthday_to_you](happy_birthday_to_you.jpg)

```python
import numpy as np
import wave
import struct

file = "happy_birthday_to_you.wav"

amplitude = 30000
duration = 1 # second
fs = 44100
num_samples = duration * fs

# Do, Re, Mi,... -> frequency Mhz
freq = {
    '&4': 174.6,
    '&5': 196,
    '&6': 220,
    '&7': 246.9,
    '1': 261.6,  # Do
    '2': 293.7,  # Re
    '3': 329.6,  # Mi
    '4': 349.2,
    '5': 392,
    ' ': 0
}

num_channels = 1
sampwidth = 2
num_frames = num_samples
comptype = "NONE"
compname = "not compressed"

music = [
    ['&5', ' ', '&5', '&6', '&5'], ['1', '&7'], ['&5', ' ', '&5', '&6', '&5'], ['2', '1'],
    ['&5', ' ', '&5', '5', '3'], ['1', '&7', '&6'], ['4', '4', '3', '1'], ['2', '1']
]

rhythm = [
    [.47, .03, .5, 1, 1], [1, 2], [.47, .03, .5, 1, 1], [1, 2],
    [.47, .03, .5, 1, 1], [1, 1, 1], [.75, .25, 1, 1], [1, 2]
]

def create_theme(theme, scale, duration):
    t = np.linspace(0, duration, num_samples*duration, endpoint=False)
    sound = amplitude * np.cos(2 * np.pi * freq[scale] * t)
    return np.append(theme, sound)

x = np.array([])
for i in range(len(music)):
    for alpha, period in zip(music[i], rhythm[i]):
        x = create_theme(x, alpha, period)

wav_file = wave.open(file, 'w')
wav_file.setparams((num_channels, sampwidth, fs, num_frames, comptype, compname))

for s in x:
    wav_file.writeframes(struct.pack('h', int(s)))

wav_file.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f1(x):
    y = np.sin(2*x)
    return y

def f2(x):
    y = np.exp(-x) * np.cos(4*x)
    return y

x = np.linspace(0, 2*np.pi, 100)
y1 = f1(x)
y2 = f2(x)

plt.plot(x, y1, x, y2)
plt.xlabel('x')
plt.ylabel('f1(x) and f2(x)')
plt.title('Plot of Two Functions')
plt.legend(['sin(2x)', 'exp(-x)cos(4x)'])
plt.grid()
plt.show()
```
